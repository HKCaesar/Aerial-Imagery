Download Maps Images

```python
from io import BytesIO
from PIL import Image
import urllib
from matplotlib.pyplot import imshow
import numpy as np
```

Image type == hybrid/satellite

```python
url = "https://maps.googleapis.com/maps/api/staticmap?center=New+York&zoom=15&size=600x600&maptype=satellite"
buffer = BytesIO(urllib.urlopen(url).read())
image = Image.open(buffer)
image.save("map_terrain.png")
```

```python
image.show()
```

```python
%matplotlib inline
imshow(np.asarray(image))
```
